```python
import matplotlib.pyplot as plt
import re
```

```python
def process_text(text):
    text = text.lower()

    words = re.findall(r'\w+', text)
    # print(words)

    word_freq = {}

    for word in words:
        if word not in word_freq.keys():
            word_freq[word] = 1
        else:
            word_freq[word] += 1
    sorted_word_freq = dict(sorted(word_freq.items(), key = lambda item : item[1],reverse = True))
    return sorted_word_freq
```

```python
def plot_histogram(word_freq):
    words = list(word_freq.keys())
    frequencies = list(word_freq.values())

    cnt = 1
    top10_words = []
    top10_frequencies = []
    for key, value in word_freq.items():
        if cnt > 10:
            break
        top10_words.append(key)
        top10_frequencies.append(value)
        cnt += 1
    
    # Plot the histogram
    plt.figure(figsize=(10, 6))
    plt.bar(words, frequencies, color='blue')
    plt.xlabel('Words')
    plt.ylabel('Frequencies')
    plt.title('Frequency Token Histogram')
    plt.xticks(rotation=90)
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.bar(top10_words, top10_frequencies, color='blue')
    plt.xlabel('Words')
    plt.ylabel('Frequencies')
    plt.title('10 most frequent tokens in description')
    plt.xticks(rotation=90)
    plt.show()
```

```python
def main():
    text = input("Enter the text: ")
    
    word_freq = process_text(text)
    
    print("Word Frequencies (sorted):")
    for word, freq in word_freq.items():
        print(f"{word}: {freq}")
    
    plot_histogram(word_freq)

if __name__ == "__main__":
    main()
```
