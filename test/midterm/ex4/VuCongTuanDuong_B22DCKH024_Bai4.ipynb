```python
import matplotlib.pyplot as plt
import re
```

```python
def process_text(text):
    text = text.lower().strip()

    words = re.split(r'[\s\.\,\!\?\:\‘\’]+', text)
    # print(words)

    word_freq = {}
    word_list = []
    word_fre_list = []
    for word in words:
        if (len(word) != 0):
            if word not in word_freq.keys():
                word_freq[word] = 1
                word_list.append(word)
            else:
                word_freq[word] += 1
    word_list.sort(key = lambda x : -word_freq[x])
    for x in word_list:
        word_fre_list.append(word_freq[x])
    return word_freq, word_list, word_fre_list
```

```python
def plot_histogram(word_list, word_fre_list):
    # Plot the histogram
    plt.figure(figsize=(10, 6))
    plt.bar(word_list, word_fre_list)
    plt.xlabel('Words')
    plt.ylabel('Frequencies')
    plt.title('Frequency Token Histogram')
    plt.xticks(rotation=90)
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.bar(word_list[:10], word_fre_list[:10])
    plt.xlabel('Words')
    plt.ylabel('Frequencies')
    plt.title('10 most frequent tokens in description')
    plt.xticks(rotation=90)
    plt.show()
```

```python
def main():
    print("INPUT")
    text = input("Nhập đoạn văn bản: ")
    
    word_freq, word_list, word_fre_list = process_text(text)
    print("OUTPUT")
    print("Số lần xuất hiện của các từ sau khi sắp xếp:")
    for i in range(len(word_list)):
        print(f"{word_list[i]}: {word_fre_list[i]}")
    plot_histogram(word_list, word_fre_list)

if __name__ == "__main__":
    main()
```
